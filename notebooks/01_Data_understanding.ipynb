```python
## Missing Values Analysis

#Missing values reduce data quality and can affect model performance.  
#This section checks whether the dataset contains any missing values.
```

```python
#check missing values in the dataset

missing_values = df.isnull().sum()
missing_df = pd.DataFrame({
    "Column": missing_values.index,
    "Missing Values": missing_values.values
})

missing_df = missing_df[missing_df["Missing Values"] > 0]

if missing_df.empty:
    print("No Missing Values Found")
else:
    display(missing_df.sort_values(by="Missing Values", ascending=False))
```

```python
### Observation

#The dataset is checked for missing values.  
#The result will help determine whether data cleaning is required before model training.
```

```python
#check duplicate records in the dataset

duplicate_records = df.duplicated().sum()
print(f"Number of duplicate records: {duplicate_records}")
```
