# Microfinance Fraud Detection using Machine Learning

# Data Preprocessing

## Objective

The objective of this notebook is to prepare the dataset for machine learning by:

- Loading the dataset
- Creating a working copy
- Removing unnecessary columns
- Handling missing values
- Handling duplicate records
- Encoding categorical variables
- Scaling numerical features
- Saving the processed dataset

```python
# ============================================================
# Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv(r"C:\Users\Dinesh\OneDrive\Desktop\Microfinance-Fraud-Detection\data\raw\microfinance_fraud_detection_dataset.csv")
print(df.shape)
df.head()
```

```python
data = df.copy()
print("Data copied successfully")
```

```python
columns_to_drop = [
    "customer_id"
]

data.drop(columns=columns_to_drop, inplace=True)

print("Remaining Shape :", data.shape)
```

```python
missing = data.isnull().sum()

missing = missing[missing > 0]

if len(missing) == 0:
    print("No Missing Values Found")
else:
    print(missing.sort_values(ascending=False))
```

```python
duplicates = data.duplicated().sum()

print("Duplicate Records :", duplicates)
```

```python
# ============================================================
# Separate Features
# ============================================================

categorical_columns = data.select_dtypes(include="object").columns.tolist()

numerical_columns = data.select_dtypes(include=["int64","float64"]).columns.tolist()

print("Categorical :", len(categorical_columns))
print("Numerical :", len(numerical_columns))
```

```python
# ============================================================
# Encode Categorical Variables
# ============================================================

label_encoders = {}

for column in categorical_columns:
    
    encoder = LabelEncoder()
    
    data[column] = encoder.fit_transform(data[column].astype(str))
    
    label_encoders[column] = encoder

print("Encoding Completed")
```

```python
import joblib
joblib.dump(label_encoders, r"C:\Users\Dinesh\OneDrive\Desktop\Microfinance-Fraud-Detection\models\label_encoders.pkl") 
print("Label Encoders Saved Successfully")
```
